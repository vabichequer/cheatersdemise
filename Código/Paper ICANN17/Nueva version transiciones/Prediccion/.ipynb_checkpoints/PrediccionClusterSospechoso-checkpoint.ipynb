### Importación de librerías y declaración de funciones

```python
# Librerías generales
import numpy as np
import pandas as pd
from __future__ import print_function

pd.set_option('display.max_columns', 30)
from IPython.core.display import display, HTML, clear_output

# Librerías ML

# Gráficos
import matplotlib.pyplot as plt
%matplotlib inline

from matplotlib import style
style.use("ggplot")
plt.rcParams["figure.figsize"] = [20,20]
```

```python
def DeepCrossValidation(dfData, dfLabels, name, estimator, param_grid, kfold=10, outfile = "print", 
                        scoring = "neg_mean_squared_error", Type = "Regression", random_state = 206, 
                        skf = None, saveScore = None):
    """
    DeepCrossValidation es una funcion que realiza doble validacion cruzada: realiza una primera
        particion sobre el conjunto de datos para testear los modelos, y para partición de entrenamiento
        realiza otra particion para elegir los mejores parametros de entrenamiento, utilizando un mallado.
        Por defecto se paraleliza a nivel de particiones, por lo que se recomienda no paralelizar a nivel
        de algoritmo, para que la ejecucion sea mas rapida.
        
    Parametros de entrada:
        - dfData: pandas dataframe con la matriz de diseno (conjunto de caracteristicas).
        - dfLabels: pandas dataframe con las etiquetas del algoritmo supervisado.
        - estimator: funcion estimador de sklearn que se desea testear con este metodo.
        - param_grid: grid de parametros para utilizar sobre la funcion estimador.
        - kfold: numero de particiones para testear y para la seleccion de mejores parametros (def = 10).
        - outfile: fichero de salida donde volcar los resultados (def "print": imprimir por pantalla).
        - scoring: metrica de scoring a utilizar para maximizar por el estimador (def: nMSE).
        - Type: tipo de problema supervisado a resolver (def: "Regression"). Si es "Classification", el
            scoring por defecto es "accuracy".
        - random_state: valor de la semilla aleatoria para reproducibilidad de resultados (def: 206).
        - skf: kfold estratificado. Si esta a None, se genera kfold estratificado. La semilla utilizada
            es random_state. Se puede pasar el tipo de kfold estratificado de manera externa para ralizar
            distintas pruebas sobre el mismo tipo de particion (def: None).
        - saveScore: estructura (lista) para indicar que se desean guardar las puntuaciones generadas
            (def: None).
            > saveScore[0] = identificadores de muestra/usuario
            > saveScore[1] = nombre del fichero csv a guardar los datos    
    """ 
    
    # Importacion de librerias
    import time
    from sklearn.model_selection import StratifiedKFold
    from sklearn.model_selection import GridSearchCV
    
    # Seleccion de metrica segun scoring
    # Si no es regresion, la metrica por defecto es accuracy (problema de clasificacion)
    if Type != "Regression":
        scoring = "accuracy"
        
    # CLASIFICACION
    if scoring == "accuracy":
        from sklearn.metrics import accuracy_score as acc
        met = acc
        
    # REGRESION
    elif scoring == "neg_mean_squared_error":
        from sklearn.metrics import mean_squared_error as mse
        met = mse
    elif scoring == "neg_mean_absolute_error":
        from sklearn.metrics import mean_absolute_error as msa
        met = msa
    elif scoring == "r2":
        from sklearn.metrics import r2_score as r2
        met = r2
    else:
        print ("Metrica no contemplada.")
        return

    
    # Inicializacion de variables
    trScores = []
    cvScores = []
    tsScores = []
    parameters = []
    i = 0

    
    # Generamos las particiones estratificadas, si no estan pasadas por argumento.
    if skf == None:
        skf = StratifiedKFold(n_splits=kfold, random_state=random_state)
        
    # Por cada conjunto de particiones
    for indTrain, indTest in skf.split(dfData, dfLabels):
        i += 1
        
        # Medimos el tiempo que tarda en realizar cada tarea
        tini = time.time()
        
        # Entrenamiento del grid con validacion cruzada. Por defecto se realizar en paralelo
        clf = GridSearchCV(estimator=estimator, param_grid=param_grid, cv=kfold, 
                           scoring = scoring, n_jobs = -1)
        
        # Entrenamos nuestro algoritmo de aprendizaje supervisado
        clf.fit(dfData[:][indTrain], dfLabels[indTrain])
        
       
        # Mejor puntuacion de validacion cruzada. Cambio de signo para las medidas negativas
        if scoring == "neg_mean_squared_error" or scoring == "neg_mean_absolute_error":
            cvScore = -clf.best_score_
        else:
            cvScore = clf.best_score_
            
        # Puntuacion de validacion cruzada
        cvScores.append(cvScore)
        
        # Entrenamos el mejor clasificador con todos los datos de la particion de train
        clf = clf.best_estimator_
        clf.fit(dfData[indTrain], dfLabels[indTrain])
        
        # Evaluacion sobre el conjunto de train
        trScore = met(dfLabels[indTrain], clf.predict(dfData[:][indTrain]))
        trScores.append(trScore)
        
        # Evaluacion sobre el conjunto de test
        Score = clf.predict(dfData[:][indTest])
        tsScore = met(dfLabels[indTest], Score)
        Score = [[x] for x in Score]
        tsScores.append(tsScore)
        
        # Para clasificacion medimos la probabilidad de pertenencia a una clase
        if Type != "Regression":
            Score = clf.predict_proba(dfData[:][indTest])
            
        # Guardar mejor parametro
        parameters.append(clf.get_params(False))

        # Guardamos en fichero, o...
        if outfile != "print":
            
            # Estructura del fichero de salida:
            # Nombre de la prueba_particion; mejores parametros de particion de test; 
            #  nombre de la puntuacion; puntuacion de train; puntuacion de validacion cruzada; 
            #  puntuacion de test; tiempo de ejecucion, fecha de realizacion de experimento.
            
            rFile = open(outfile, 'a')
            rFile.write(name+"_"+"%02d"%(i)+';'+str(clf.get_params(False))+";"+str(scoring)+";"
                        +str(trScore)+";"+str(cvScore)+";"+str(tsScore)+";"+str(time.time()-tini)+";"
                        +time.ctime()+"\n")
            rFile.close()
            
            
            #saveScore[0] = identificadores de muestra/usuario
            #saveScore[1] = nombre del fichero csv a guardar los datos
            
            # Estructura del fichero de puntuaciones:
            # Nombre de la prueba_particion; id de la muestra/usuario; label; [puntuacion].
            if saveScore != None:
                rFile = open(saveScore[1], 'a')
                for ind in range(len(indTest)):
                    rFile.write(name+"_"+"%02d"%(i)+';'+saveScore[0][indTest[ind]]+";"
                                +str(dfLabels[indTest[ind]])+";["+', '.join(map(str, Score[ind]))+"]\n")
                rFile.close()
            
            print (name+"_"+"%02d"%(i), "- Tiempo de proceso [s]:","%.03f"%(time.time()-tini))

        # Mostramos por pantalla
        else:
            print (name+"_"+"%02d"%(i), "- Tiempo de proceso [s]:","%.03f"%(time.time()-tini))
            print (clf)
            print
            print ("Puntuación de entrenamiento:", trScore)
            print ("Puntuación de validación:", cvScore)
            print ("Puntuación de test:", tsScore)
            print

    # Se devuelven las puntuaciones de train, validacion y test en una lista, una lista con los
    #  mejores parametros de cada validacion cruzada y el kfold estratificado utilizado.
    return [trScores, cvScores, tsScores], parameters, skf
```

```python
def medianParameters(paramList):
    """
    Calcula los parametros mas frecuentes de una lista de parametros. La lista puede dar los parametros
        en formato de texto o en formato diccionario.
    Devuelve una lista con la frecuencia de aparicion del conjunto de parametros y una lista con el
        conjunto de parametros (en el caso en el que haya colision habra mas de un elemento en ambas 
        listas).
    """
    
    # Transformamos el tipo si nos pasan una lista de diccionarios a lista de str
    if type(paramList[0]) == dict:
        paramList = [str(x) for x in paramList]
        
    # Generamos una sublista consistente en la frecuencia de cada palabra y la palabra de la lista
    # Esta lista esta ordenada de mas a menos frecuente
    from collections import Counter
    mostCommon = [[wc, w] for w, wc in Counter(paramList).most_common(len(paramList))]
    
    maxim = 0
    freq = []
    param = []
    
    # Buscamos el elemento mas frecuente
    for w in mostCommon:
        
        # Si la frecuencia es mayor o igual anadimos dicha palabra con su frecuencia a las listas
        if w[0] >= maxim:
            maxim = w[0]
            param.append(w[1])
            freq.append("%.2f"%(float(maxim)/len(paramList)))
            
        # Una vez no haya palabras mas frecuentes, salimos del for
        else:
            break
            
    # Devolvemos la frecuencia de cada conjunto de parametros con dicho conjunto de parametros
    return freq, param
```

```python
def evaluateResults(scores, paramList, dec=3, save = None, console = True):
    
    """
    Esta funcion permite guardar los resultados de la validacion cruzada en un fichero CSV ademas
        de imprimirlos por pantalla (con media y std). Tambien guarda en el mismo fichero los conjuntos
        de parametros mas frecuentes y los muestra por pantalla.
    
    Argumentos de entrada:
        - scores: lista con tres listas, con las puntuaciones de validacion cruzada de train, validacion
            y test.
        - paramList: lista de mejores parametros de entrenamiento resultantes de la validacion cruzada.
        - dec: precision (posiciones decimales) de las puntuaciones medias y su desviacion (def: 3).
        - save: fichero de salida donde guardar los resultados. (def: None)
        - console: imprimir los resultados por consola (def: True)
    """
    import numpy as np
    
    # Abrimos el fichero
    if save != None:
        rFile = open(save, 'a')
    
    # Puntuaciones de tr, cv, ts
    code = "%.0"+str(dec)+"f"
    
    # Variables de resultados
    name = ["Train", "CrossVal", "Test"]
    i = 0
    
    for s in scores:
        
        # Calculamos la media y std de cada conjunto de resultados
        mean = np.mean(s)
        std = np.std(s)
        
        # Guardamos los resultados en formato lista de python
        if save != None:
            rFile.write("["+code%(mean)+", "+code%(std)+"];")
        
        # Imprime por pantalla la evaluacion de las puntuaciones de train, validacion y test.
        if console:
            print (name[i],"Score:\t",code%(mean)+" +- "+code%(std))
            
        i += 1
            
    # Conjunto de parametros mas frecuente
    freq, param = medianParameters(paramList)
    
    # Guardamos los resultados en formato lista de python
    if save != None:
        rFile.write(str(freq)+";"+str(param)+"\n")
        rFile.close()
    
    # Imprimimos por pantalla los parametros mas frecuentes.
    if console:
        print("\n")
        print(freq,param)
        
```

### Predicción de abandono

```python
# MODIFICAR
nombre = "PrediccionClusterSospechoso"
folder = "./Resultados_"+nombre+"/"
outfile = folder+"pruebasPrediccion"+nombre+".csv"
outfile2 = folder+"resultadosPrediccion"+nombre+".csv"

kfold = 10
seed = 206
rseed = [seed]


import os
if not os.path.exists(folder):
    os.makedirs(folder)
```

```python
# MODIFICAR
dfData = pd.read_csv("Prediccion_suspect_Ed1.csv", sep = "|")

# Aleatorizacion de los datos de entrada
import random
random.seed(seed)
dfData = dfData.sample(frac = 1, random_state = seed)


labels = np.array(dfData.suspect)
users = np.array(dfData[dfData.columns[0]].apply(str))
dfData = dfData.drop([dfData.columns[0],"suspect"],axis=1)
df = np.array(dfData)
```

```python
# Logistic Regression
from sklearn.linear_model import LogisticRegression as LogReg

name = "LogisticRegression_"
C = [0.1, 1.0, 10]
max_iter = [100, 200]

grid = [{'C': C,
         'max_iter': max_iter,

         'n_jobs': [1],
         'random_state': rseed}]

sc, par, skf = DeepCrossValidation(df, labels, name = name, estimator = LogReg(), 
                random_state = seed, param_grid=grid, kfold=kfold, Type = "Classification", 
                outfile = outfile, saveScore = [users, folder+"scores"+name+".csv"])

print("\n")
evaluateResults(sc, par, save = outfile)
evaluateResults(sc, par, save = outfile2, console = False)
print("\n\n")
```

```python
# RandomForest Classifier
from sklearn.ensemble import RandomForestClassifier as RFC

name = "RandomForestClassifier_"

n_estimators = [500]
max_features = ["sqrt", "log2"]
criterion = ["gini"]

grid = [{'n_estimators': n_estimators,
         'max_features': max_features,
         'criterion': criterion,
         'random_state': rseed}]

sc, par, skf = DeepCrossValidation(df, labels, name = name, estimator = RFC(), 
                random_state = seed, param_grid=grid, kfold=kfold, Type = "Classification", 
                outfile = outfile, saveScore = [users, folder+"scores"+name+".csv"], skf=skf)
print("\n")
evaluateResults(sc, par, save = outfile)
evaluateResults(sc, par, save = outfile2, console = False)
print("\n\n")
```

```python
# XGBoost Classifier
from xgboost.sklearn import XGBClassifier as XGBC

name = "XGBoostClassifier_"

n_estimators = [100,200,300]
max_depth = [5,7]
learning_rate = [.1]

grid = [{'n_estimators': n_estimators,
         'max_depth': max_depth,
         'learning_rate': learning_rate,
         'seed': rseed}]

sc, par, skf = DeepCrossValidation(df, labels, name = name, estimator = XGBC(), 
                random_state = seed, param_grid=grid, kfold=kfold, Type = "Classification", 
                outfile = outfile, saveScore = [users, folder+"scores"+name+".csv"], skf=skf)
print("\n")
evaluateResults(sc, par, save = outfile)
evaluateResults(sc, par, save = outfile2, console = False)
print("\n\n")
```

```python
# Decision Tree Classifier
from sklearn.tree import DecisionTreeClassifier as DTC
name = "DecisionTreeClassifier_"

criterion = ['gini', 'entropy']
max_depth = [5,7,9,11,21,31]
max_features = ['auto', 'sqrt', 'log2']

grid = [{'criterion': criterion,
         'max_depth': max_depth,
         'max_features': max_features,
         'random_state': rseed}]

sc, par, skf = DeepCrossValidation(df, labels, name = name, estimator = DTC(), 
                random_state = seed, param_grid=grid, kfold=kfold, Type = "Classification", 
                outfile = outfile, saveScore = [users, folder+"scores"+name+".csv"], skf=skf)
print("\n")
evaluateResults(sc, par, save = outfile)
evaluateResults(sc, par, save = outfile2, console = False)
print("\n\n")
```

```python
# NeuralNetwork Classifier
from sklearn.neural_network import MLPClassifier as MLP

name = "MLPClassifier_"

hidden_layer_sizes = [(50,), (100,), (50, 10), (50, 50), (100, 10), (100, 50)]
alpha = [.0001, .001, .01, .1, 1, 10]
max_iter = [200, 300, 500]

grid = [{'hidden_layer_sizes': hidden_layer_sizes,
         'alpha': alpha,
         'max_iter': max_iter,
         'random_state': rseed}]

sc, par, skf = DeepCrossValidation(df, labels, name = name, estimator = MLP(), 
                random_state = seed, param_grid=grid, kfold=kfold, Type = "Classification", 
                outfile = outfile, saveScore = [users, folder+"scores"+name+".csv"], skf=skf)
print("\n")
evaluateResults(sc, par, save = outfile)
evaluateResults(sc, par, save = outfile2, console = False)
print("\n\n")
```

```python
from sklearn.tree import _tree

def leaf_depths(tree, node_id = 0):
    
    '''
    tree.children_left and tree.children_right store ids
    of left and right chidren of a given node
    '''
    left_child = tree.children_left[node_id]
    right_child = tree.children_right[node_id]

    '''
    If a given node is terminal, 
    both left and right children are set to _tree.TREE_LEAF
    '''
    if left_child == _tree.TREE_LEAF:

        '''
        Set depth of terminal nodes to 0
        '''
        depths = np.array([0])

    else:

        '''
        Get depths of left and right children and
        increment them by 1
        '''
        left_depths = leaf_depths(tree, left_child) + 1
        right_depths = leaf_depths(tree, right_child) + 1

        depths = np.append(left_depths, right_depths)
 
    return depths

def leaf_samples(tree, node_id = 0):
    
    left_child = tree.children_left[node_id]
    right_child = tree.children_right[node_id]

    if left_child == _tree.TREE_LEAF:

        samples = np.array([tree.n_node_samples[node_id]])

    else:

        left_samples = leaf_samples(tree, left_child)
        right_samples = leaf_samples(tree, right_child)

        samples = np.append(left_samples, right_samples)

    return samples

def draw_tree(ensemble, tree_id=0):

    import matplotlib.pyplot as plt
    
    plt.figure(figsize=(8,8))
    plt.subplot(211)
    
    tree = ensemble.estimators_[tree_id].tree_

    depths = leaf_depths(tree)
    plt.hist(depths, histtype='step', color='#9933ff', 
              bins=range(min(depths), max(depths)+1))

    plt.xlabel("Depth of leaf nodes (tree %s)" % tree_id)
    
    plt.subplot(212)
    
    samples = leaf_samples(tree)
    plt.hist(samples, histtype='step', color='#3399ff', 
              bins=range(min(samples), max(samples)+1))
    
    plt.xlabel("Number of samples in leaf nodes (tree %s)" % tree_id)
   
    plt.show()
    
def draw_ensemble(ensemble):

    plt.figure(figsize=(8,8))
    plt.subplot(211)

    depths_all = np.array([], dtype=int)

    for x in ensemble.estimators_:
        tree = x.tree_
        depths = leaf_depths(tree)
        depths_all = np.append(depths_all, depths)
        plt.hist(depths, histtype='step', color='#ddaaff', 
              bins=range(min(depths), max(depths)+1))

    plt.hist(depths_all, histtype='step', color='#9933ff', 
          bins=range(min(depths_all), max(depths_all)+1), 
          weights=np.ones(len(depths_all))/len(ensemble.estimators_), 
          linewidth=2)
    plt.xlabel("Depth of leaf nodes")

    samples_all = np.array([], dtype=int)

    plt.subplot(212)

    for x in ensemble.estimators_:
        tree = x.tree_
        samples = leaf_samples(tree)
        samples_all = np.append(samples_all, samples)
        plt.hist(samples, histtype='step', color='#aaddff', 
              bins=range(min(samples), max(samples)+1))

    plt.hist(samples_all, histtype='step', color='#3399ff', 
          bins=range(min(samples_all), max(samples_all)+1), 
          weights=np.ones(len(samples_all))/len(ensemble.estimators_), 
          linewidth=2)
    plt.xlabel("Number of samples in leaf nodes")

    plt.show()
```
