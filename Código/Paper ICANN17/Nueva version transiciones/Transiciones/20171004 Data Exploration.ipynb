This Notebook computes and plots the distribution of the answers for each question in MBT, and also plots the transition matrix between pretest and post test.

```python
# Plot in the notebook
import matplotlib.pyplot as plt
%matplotlib inline

import pandas as pd
```

```python
# answers = [' ', 'A', 'B', 'C', 'D', 'E', 'AB', 'AC', 'AD', 'AE', 'BC', 'BD', 'BE', 'CD', 'CE', 'DE']

correct = ['B', 'D', 'E', 'C',
          'A', 'C', 'C', 'D',
          'A', 'E', 'E', 'C',
          'B', 'B', 'E', 'A',
          'D', 'B', 'C', 'C',
          'A', 'B', 'D', 'A',
          'A', 'E']

# We consider that every answer out of the set A-E is not valid (X) -combination of letters or no answer-.
valid = ['A', 'B', 'C', 'D', 'E']
possible = valid + ['X']

# Initialization of the Questions structure: #Question x kindOfTest x Answer x value (1)
Questions = []

# Files to obtain data
#files = ['proc_data/MBT2005pre.csv', 'proc_data/MBT2007pre.csv', 'proc_data/MBT2008pre.csv']
#files = ['proc_data/MBT2005post.csv', 'proc_data/MBT2007post.csv', 'proc_data/MBT2008post.csv']

#files = ['proc_data/MBT2005pre.csv', 'proc_data/MBT2005post.csv', 'proc_data/MBT2005-2009re.csv']
#files = ['proc_data/MBT2007pre.csv', 'proc_data/MBT2007post.csv']
#files = ['proc_data/MBT2008pre.csv', 'proc_data/MBT2008post.csv']

#files = ['proc_data/MBT2005postInter.csv', 'proc_data/MBT2007postInter.csv', 'proc_data/MBT2008postInter.csv']
#files = ['proc_data/MBT2005preInter.csv', 'proc_data/MBT2007preInter.csv', 'proc_data/MBT2008preInter.csv'] 
files = ['proc_data/AllPretestInter.csv', 'proc_data/AllPosttestInter.csv']
#files = ['proc_data/AllPretest.csv', 'proc_data/AllPosttest.csv']



# List of dfs loaded
lstDF = []
```

```python
# for over the files
for inFileName in files:
    
    # Read the files. ID as an index.
    df = pd.read_csv(inFileName, index_col = 0)

    # Change the name of the columns by the number of each question.
    df.columns=list(map(str,range(1,len(df.columns)+1)))
    
    # If the answer is not valid, we set it as 'X'
    df = df.applymap(lambda x: 'X' if x not in valid else x)

    # Save the df in the list
    lstDF.append(df)

    # Print the shape of the dataframe (#Students x #Answers)
    print(inFileName, '\t', df.shape)
    
    # for over the questions
    for q in range(len(df.columns)):
        
        # Get the count of each answer for each question.
        # Normalize by the number of students.
        values = (df[str(q+1)].value_counts()/df.shape[0])

        # List of counts of possible answers, in the the following order: A, B, C, D, E, X.
        lstAnsw = []
        
        # for over possible answers
        for ans in possible:
            # an answer sometimes does not appear in a question
            # (ex: none answers 'A' for a question), launching an exception.
            try:
                lstAnsw.append(values[ans])
            # if that answer does not appear, its frequency is equal to 0.
            except:
                lstAnsw.append(0.)
                
        # the first time executing this function Questions is void,
        # launching an exception if accessing Questions[q].
        try:
            Questions[q].append(lstAnsw)
        # if exception, create a new list of answers for that question.
        except:
            Questions.append([lstAnsw])
    
```

### Bar diagram (distribution of responses)

```python
# This function was implemented enterely out of MIT by Angel Perez Lemonche
def barPlot(data, width = .30, tags = [], title = '', labels = [], colors = [], edgecolor = [],
            legend = [], ax=[], plotValues = []):
    '''
    barPlot plots a bar diagram with one or more group of bars. It also plots the std as a vertical
        line if provided.
        
    Input arguments:
        - data: data is the information to plot and has the following structure: groupOfBars x 
            numberOfTicks x data+std. The information is provided as a list of lists of lists (if std).
        - width: with of the bars. Caution, the bars can overlap if width is a large number.
        - tags: list of names of the xticks.
        - title: title of the figure.
        - labels: list with the name of the [xlabel, ylabel].
        - colors: color of the bars. Recomendation for three bars: ['b', 'r', 'g'] for colorful bars, or
            ['lightgrey','darkgrey','dimgrey'] for grayscale bars.
        - edgecolor: color of the edge of the bar. It is recomended for the grayscale colors to set
            edgecolor to 'black' or 'gray', in order to visualize better.
        - legend: legend of the figure. Modify code if not willing to use it. Fontsize 9.
        - ax: for subplots, ax is the axes of the figure. If not provided, a new figure is created.
            This function returns the axes for further modifications outside itself.
        - plotValues: if provided, plots the values of the bars. It is a list of three dimentions:
            [xDeviation, yDeviation, rotation]. The values are plotted in black, fontsize 9.
            
        Note: the location of the legend is hardcoded in this function. To change it, just change the 
            code at the end of the function.
    '''
        
    # number of group of bars
    N = len(data)
    
    # list of ticks
    import numpy as np
    ind = np.arange(len(data[0]))
    
    # create a new figure if ax is void
    if ax == []:
        fig, ax = plt.subplots()

        
        
    # test for labels
    if len(labels)==2:
        ax.set_xlabel(labels[0])
        ax.set_ylabel(labels[1])
        
    # title
    ax.set_title(title)
    
    # xticks location
    ax.set_xticks(ind)
    
    # xticks names
    if len(tags)==0:
        tags = range(1,N+1)
    ax.set_xticklabels(tags)
    
    # grid
    ax.grid(True)
    
    
    
    # graph counts along group of bars (GOBars)
    graph = 0
    
    for GOBars in data:
        
        # init. of means and std
        bar_means = []
        bar_std = []
        
        # values moves along GOBars
        for values in GOBars:
                        
            # if values is a list
            try:
                plotStd = bool(len(values)-1)
                bar_means.append(values[0])
                
                # if has std
                if plotStd:
                    bar_std.append(values[1])
                    
            # if values is a float
            except:
                plotStd = False
                bar_means.append(values)
                
        # if has std to be plotted
        if plotStd:
            ax.bar(ind+graph*width-width*(N-1)/2, bar_means, width, color = colors[graph], yerr = bar_std, 
                   edgecolor = edgecolor)
        else:
            ax.bar(ind+graph*width-width*(N-1)/2, bar_means, width, color = colors[graph], edgecolor = edgecolor)
        
        # plot the value of the bar
        if len(plotValues)>2:
            for v in ind:
                ax.text(v+graph*width-width*(N-1)/2 + plotValues[0], plotValues[1], "%.2f%%"%(bar_means[v]*100), 
                        color='k', rotation=plotValues[2], size=9)
                
        # update graph value
        graph+=1
        
        
        
    # legend
    # Careful: legend location depends on the plot we are doing
    ax.legend(legend, bbox_to_anchor = (4.8/6, 1), loc = 2, fontsize = 9)


        
    # if its a new figure
    if ax == []:
        return ax, fig
    
    # if was a previous figure
    else:
        return ax
    
```

```python
# Plotting the results

# siz is just a variable to control the size of the plot.
# the smaller, the better, but probably there will be overlap of figures.
siz = 4

# Create a new figure (size in width x height)
fig = plt.figure(figsize = (siz*3, siz*7))

# for over Questions
for q in range(len(Questions)):
    
    # Since there are 26 questions, we plot 9x3 (one of them will be void)
    ax = plt.subplot(9,3,q+1)
    
    # Call to barplot. Read the arguments of the function for further information.
    barPlot(Questions[q], width = .30, tags = possible, title = 'Question ' + str(q+1),
            labels = ['Answer','[%] Students'], colors = ['lightgrey','dimgrey'], #'darkgrey',
            edgecolor = 'gray', legend = ['preAll','postAll','pre2008'], ax = ax, plotValues = [-0.08,0.2,90])
    
    # Another legend ['post2005','post2007','post2008']darkgrey
    
    # Set the y limit between 0 and 1 for every plot.
    ax.set_ylim([0,1])
    
# Tight layout to avoid overlapping among figures.
fig.tight_layout()

# Save the figure in vectorial format (eps)
fig.savefig('AnswerDistribution%_prepostAll_Inter(05-10).eps', format='eps')
```

### Transition Matrices

Plot the transition matrices (from pre test to post test). Also can be plotted normalized.

```python
def transMatrix(df, possibleValues):
    '''
    Computes a cumulated transition matrix given a dataframe with two columns
        with numeric values (starting in 0) and a list with the possible number
        of values it can take.
    '''
    
    import numpy as np
    
    # Initialization of the matrix
    M = np.zeros((len(possibleValues), len(possibleValues)))
    
    # for each row in the dataframe (entry or register)
    for row in range(len(df)):
        
        # Add 1 in the possition given by the first and the second row
        M[df[df.columns[0]][row]][df[df.columns[1]][row]] += 1
        
    # return with int type
    return M.astype(int)
```

```python
# Create the pre and post dataframe, taking the elements in the intersection of boths sets
prepost = pd.concat([lstDF[0], lstDF[1]], axis=1, join='inner')

# Change the name of the columns: 001 for the pretest first question, 110 for the post test tenth question...
prepost.columns = ['0'+'%02d'%(x) for x in range(1,lstDF[0]
                        .shape[1]+1)] + ['1'+'%02d'%(x) for x in range(1,lstDF[1].shape[1]+1)]

# Conversion from characters (A, B, C...) to numbers (0, 1, 2...).
prepost = prepost.applymap(lambda x: possible.index(x))
```

```python
def plotTransMatrix(cm, classes, normalize = False, title = 'Transition Matrix', cmap = plt.cm.Greys, Sum = True, perc = True):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting 'normalize = True'.
    """
    
    import numpy as np
    import itertools
    perc = (perc and normalize)*99 + 1
    
    # If normalize flag is true, we normalize in the 0 axis
    if normalize:
        cm = cm.astype('float') / sum(cm.sum(axis=1)[:, np.newaxis])
        # having done that, we may obtain 0/0 inconsistences. 
        # we define 0/0 as 0
        cm = np.nan_to_num(cm)
        title = 'Normalized '+title

    # Add sum of the answers
    if Sum:
        # Initialization of the matrix
        black = False


        if black:
            c = cm.max()
        else:
            c = 0
        row = c*np.ones(np.shape(cm)[1])
        cm2 = np.vstack((cm, row))
        cm3 = np.transpose(cm2)
        row = c*np.ones(np.shape(cm3)[1])
        cm4 = np.vstack((cm3, row))
        cm = np.transpose(cm4)
        

    # Plot the confusion matrix
    plt.imshow(cm*perc, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    
    # Set the ticks for both axes
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes)
    plt.yticks(tick_marks, classes)

    # Format of the printing 
    #fmt = '.2f' if normalize else 'd'
    fmt = '.0f'


    
    # Threshold to change the color
    thresh = cm.max() / 2.
    
    # Set the text over each cell
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
                      
        # Print the sum of all rows and columns
        if i == len(classes) and j == len(classes) and Sum:
            text = ' '
            #text = (format(cm.sum()*perc, fmt))
            fw = 'normal'
        elif j == len(classes) and Sum:
            text = format(cm[i,0:len(classes)].sum()*perc, fmt)
            fw ='bold'
        elif i == len(classes) and Sum:
            text = format(cm[0:len(classes),j].sum()*perc, fmt)
            fw ='bold'
        else:
            text = format(cm[i, j]*perc, fmt)
            fw = 'normal'
            
            
        plt.text(j, i, text, horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black", fontweight=fw)

    # Set tight layout and x and y labels
    plt.tight_layout()
    plt.ylabel('Pre Test')
    plt.xlabel('Post Test')
```

```python
def moreProbableIncorrectPostTest(TM, correctPos, possible, question):
    import numpy as np
    # Delete correct in post test
    total = TM.sum()
    TM[:,correctPos] = -1
    val = np.sort(TM.flatten())
    val = np.unique(val[val>0])
    lst = []
    for i in val:
        x,y = np.where(TM==i)
        for j in range(len(x)):
            lst.append((i/total, 'Q'+str(question)+possible[x[j]]+'2'+possible[y[j]]))
    return lst
```

```python
# Create a new figure (size in width x height)
fig = plt.figure(figsize = (siz*3, siz*7))

# Normalization of the matrix (0 axis)
norm = True

# Number of questions of the test
NoQuestions = lstDF[0].shape[1]

ProbWrongAns = []
for q in range(1,NoQuestions+1):
    # Since there are 26 questions, we plot 9x3 (one of them will be void)
    ax = plt.subplot(9,3,q)
    
    # Compute and plot the transition matrix
    TM = transMatrix(prepost[['0'+'%02d'%(q),'1'+'%02d'%(q)]], possible)
    plotTransMatrix(TM, possible, norm, title = 'Transition Matrix Q'+str(q))
    
    correctPos = possible.index(correct[q-1])
    ProbWrongAns.append(moreProbableIncorrectPostTest(TM, correctPos, possible, q))


# Save the figure in vectorial format (eps)
fig.savefig('NormTransitionMatrixWithCummulated_prepostAll(05-10).eps', format='eps')

ProbWrongAns = [item for sublist in ProbWrongAns for item in sublist]
ProbWrongAns = sorted(ProbWrongAns, key = lambda x: x[0], reverse = True)
```
